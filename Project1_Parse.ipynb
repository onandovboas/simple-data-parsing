# Projeto 1: Extração e Limpeza de dados JSON

Projeto simula a captação de dados de uma campanha de uma escola de inglês.

O código realiza o tratamento dos dados para o Python.

```python
dados_json_brutos = """
[
  {
    "nome_completo": "   cArlos eDuardo",
    "idade": 15,
    "email": "carlos.edu@gmail.com",
    "nivel_sugerido": "B1"
  },
  {
    "nome_completo": "MARIA SILVA",
    "idade": "vinte",
    "email": "mariasilva@ mms.com.br",
    "nivel_sugerido": "A2"
  },
  {
    "nome_completo": "joão pedro",
    "idade": 22,
    "email": "joao.pedro#hotmail.com",
    "nivel_sugerido": "C1"
  },
  {
    "nome_completo": "Ana Souza",
    "email": "ana.souza@gmail.com",
    "nivel_sugerido": "B2"
  }
]
"""
```

```python
import json

dados_json = json.loads(dados_json_brutos)
print(dados_json)
```

## Escrevendo um CSV com os dados em dicionário


```python
import csv


with open("dados_limpos.csv", "w", newline='') as csvfile:
  fieldnames = ["nome_completo", "idade", "email", "nivel_sugerido"]
  writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

  writer.writeheader()
  writer.writerows(dados_json)
```
